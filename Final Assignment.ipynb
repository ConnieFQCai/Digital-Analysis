```python
#Retrieving tweets
#The first function allows us to get all the individual tweets from the dataset.
#It uses `data` and looks for certain information in it.
#When applied to a dataframe, we can obtain a new dataframe with a individual tweet as the unit of analysis.
#Credited to the University of Amsterdam's Digital Analysis course.

def get_public_metrics(row):
    if 'public_metrics' in row.keys():
        if type(row['public_metrics']) == dict:
            for key, value in row['public_metrics'].items():
                row['metric_' + str(key)] = value
    return row

def get_tweets(df):
    if 'data' not in df.columns:
        return None
    results = pd.DataFrame()
    for item in df['data'].values.tolist():
        results = pd.concat([results, pd.DataFrame(item)])
        
    results = results.apply(get_public_metrics, axis=1)
        
    results = results.reset_index()
    del results['index']
        
    return results
```

```python
#move get_tweets(df) into  tweets

tweets = get_tweets(df)
```

#### How often does the words related to period show up in the dataset?

```python
#count_periods() function counts the # of times that the words related to 'period' show up in the datatset.
#f creates a f-string which allows you to include variables in strings.
#where do I put the return?

def count_periods(tweets):

    period_counter = 0
    not_period_counter = 0

    for i in range(len(tweets)):
        if any(word in tweets['text'][i].lower() for word in ['period','menstruation','menstrual']):
            period_counter += 1
        else:
            not_period_counter += 1

    return period_counter, not_period_counter        
    
 #   print(f'Period is in {period_counter} tweets, and period is not in {not_period_counter} tweets')
```

```python
#let's try this function!

period_counter, not_period_counter = count_periods(tweets)

print(f'Period is in {period_counter} tweets, and period is not in {not_period_counter} tweets')
```

#### What is the most common source of tweet?

```python
#count_source() function generates the most common source of tweets.

def count_source(tweets):

    tweets['source_counter'] = 1

    new_source = tweets.groupby('source').agg({'source_counter': 'sum'})

#sort_values() with an additional parameter ascending which makes sure it sorts from highest to lowest.

    new_source = new_source.sort_values('source_counter',ascending=False)

    return new_source.head()
    
```

```python
#let's try this function!

count_source(tweets)
```

#### What is the tweet with the most characters?

```python
#tweet_most_characters() generates the tweet with the most characters.
#where do I put return?

def tweet_most_characters(tweets):

    characters = 0

    for tweet in tweets['text']:
        xcharacters = len(tweet)
        if xcharacters > characters:
            characters = xcharacters
            twt = tweet
    
    print ("The tweet with the most characters has:", characters, ",and the entire tweet is:",twt)
```

```python
tweet_most_characters(tweets)
```

#### What is the tweet with most words & its author id?

```python
#tweets with the most words & author id

def longest_tweet_finder(tweets):
    words = 0

    for i in range(len(tweets)):
        y = len(tweets['text'][i].split(" "))
        if y > words:
            words = y
            twt = tweets['text'][i]
            author = tweets['author_id'][i]
    return words, twt, author
```

```python
longest_tweet_finder(tweets)
```

#### What is the longest word in tweets?

```python
#longest word in tweets
def longest_word_finder(tweets):
    longword = 0

    for tweet in tweets['text']:
        a = tweet.split(" ")
        for word in a:
            b = len(word)
            if b > longword:
                longword = b
                longest_word = word

    return longest_word
```

#### What are the top dates that had the most tweets?

```python
#7 days, but I want to categorize number of tweets per day
#example data: 2022-09-18T18:02:56.000Z. This is an object: combo of strings & ints
#can be split into two parts with T

#get_date() function splits the date at T.

def get_date(datestamp):
    return datestamp.split("T")[0]

#let's categorize the dates
#create a new column called 'dummydate'

tweets['dummydate'] = 1


tweets['date'] = tweets['created_at'].apply(lambda datestamp: get_date(datestamp))
new_tweets = tweets.groupby('date').agg({'dummydate': 'sum'})
new_tweets = new_tweets.sort_values('dummydate',ascending=False)

#head() function displays by default first n=5 rows.

new_tweets.head()
```
